```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[-1]
df=pd.read_csv("Student_my.csv")
```

```python
# How to know whether the data has missing values?
df.info()
```

```python
df.isnull().sum()
```

```python
# Deleting the columns with missing data
df.dropna(axis=1)
```

```python
# Deleting the rows with missing data
df.dropna(axis=0)
```

```python
# Deleting the row which contain null value in Java column
# Deleting the rows for a specific column with missing data
#how='any' (default): Drops a row if any of the specified columns have NaN.
#how='all':Drops a row only if all of the specified columns have NaN.

df.dropna(subset=['Java'],how='any')
```

```python
# Filling the missing data with a value 0
df.fillna(0)
```

```python
# Filling the missing data with a value 0 of C++ column

df['C++'].fillna(0)
```

```python
df.fillna(value = {'Python' : 11 },inplace=True)
df
```

```python
# Fill in the missing value for row number 3 and column name C
df.at[3,'C']=100
df
```

```python
# Fill the missing value of Java column with mean value
df['Java'].fillna(df['Java'].mean())
df
```

```python
data=df['C'].fillna(df['C'].mean())
data
```

```python
# Replacing with the value from the previous row or the next row
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
# Use different interpolation methods to fill the null values
df.interpolate(method='nearest')
```

```python
df.interpolate(method='linear')
```

```python
df.interpolate(method='quadratic')
```
