```python
#Read .xlsx file from drive
import pandas as pd
path="Datavisu.xlsx"
df=pd.read_excel(path,index_col=0)
print(df.head())
```

```python
# Draw the dot plot between car name and age (first 10 cars)
import matplotlib.pyplot as plt
plt.plot(df['Name'].head(10),df['Age'].head(10),linewidth=0,marker='o',color='blue')#-->Linewidth=0(dot plot)
# plt.plot(df['Name'].head(10),df['Age'].head(10),'ro--')
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Cars Vs Age")
```

```python
# Draw the line plot between age and price. Observe the trend line.
# plt.plot(df['Age'].head(10),df['Price'].head(10),linewidth=1,marker='o',color='blue')#-->Linewidth=1(Line plot)
t=df.sort_values(by='Age',ascending=True)
plt.plot(t["Age"].head(10),t['Price'].head(10),linewidth=1,marker='o',color='blue')
```

```python
# Draw the box plot to show the statistical summary of Age column and verify with describe()
print(df['Age'].describe())
plt.boxplot(df['Age'])
plt.title("BoxPlot")
```

```python
# Draw the scatter plot between Age and Maintainance cost. Observe the correlation
plt.scatter(df['Age'],df['Maintainance'],c='red')
plt.title("Age")
```

```python
# Draw the scatter plot between Age and Price. Observe the correlation
x=df['Age'].head(10)
y=df['Price'].head(10)
plt.scatter(x,y,c='red')
plt.title("Age and Price")
```

```python
# Draw the scatter plot between Age and Weight. Observe the correlation
x=df['Age'].head(10)
y=df['Weight'].head(10)
plt.scatter(x,y,c='red')
plt.title("Age and Weight")
```

```python
# Draw a bar chart with Fuel Type and its count (use different colors for each fuel type)
data=df['FuelType'].value_counts()
print(data)
plt.bar(data.index,data.values,color={'red','blue','green'})
plt.title("Bar")
```

```python
# Draw a horizontal bar chart for Fuel Type and Price
data=df['FuelType'].value_counts()
print(data)
plt.barh(data.index,data.values,color={'red','blue','green'})#For horizontal-->barh
plt.title("Barh")
```

```python
# Draw a comparitive bar chart for Price and New_price against each car name (first 10 cars)
import numpy as np
N=10
ind=np.arange(N)
width=0.25
xvals=df['Price'].head(10)
bar1=plt.bar(ind,xvals,width,color='r')

yvals=df['New_Price'].head(10)
bar2=plt.bar(ind+width,yvals,width,color='g')

plt.xticks(ind+width,df['Name'].head(10));
plt.legend((bar1,bar2),('Price','New_Price'))
plt.title("CArs names")
```

```python
# Draw a stacked bar chart for Price and New_price against the car name (first 10 cars)
plt.bar(df['Name'].head(10),df['Price'].head(10))
plt.bar(df['Name'].head(10),df['New_Price'].head(10),bottom=df['Price'].head(10))
```

```python
# Draw a pie chart with HP and its count
plt.pie(data.values,labels=data.index,autopct='%1.2f%%');
plt.title("Pie")
```

```python
# Draw the histogram plot for Age column
plt.hist(df['Age'],edgecolor='white')

# plt.hist(df['Age'],edgecolor='white')
```

```python
# Draw the histogram plot for Age column with bin value
plt.hist(df['Age'],edgecolor='white',bins=5);
```

```python
# Draw the histogram plot for Age column with PDF
plt.hist(df['Age'],edgecolor='white',density=True)#==>density=True-->All with 0 to 1;
```
